```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#linear regression without library
dataset=pd.read_csv('Salary.csv')
dataset
```

```python
#data processing
x=dataset.iloc[:, :-1].values
y=dataset.iloc[:, 1].values
```

```python
# for  simple linear regression without library 
n=np.size(x)
mean1=np.mean(x)
mean2=np.mean(y)
ss_xy=np.sum(y*x)-n*mean2*mean1   #sum of cross deviation of y and x ...or (x-x')(y-y')
ss_xx=np.sum(x*x)-n*mean1*mean1   #sum of squared deviations of x ...or (x-x')^2
b1=ss_xy/ss_xx
b0=mean2-b1*mean1
```

```python
def reg_line(x,y):
    plt.scatter(x,y, color='m', marker="o", s=30)
    y_pred=b0+b1*x
    print(y_pred)
    plt.plot(x, y_pred, color='g')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()


reg_line(x,y)
```
