```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
dataset=pd.read_csv('heart_failure_clinical_records_dataset.csv')
dataset
```

```python
x=dataset.iloc[:, :-1].values
y=dataset.iloc[:, 12].values
```

```python
x
```

```python
y
```

```python
#splitting data
x_train, x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
print("shape of x_train", x_train.shape)
print("shape of y_train", y_train.shape)
```

```python
regressor=LinearRegression()
```

```python
# fitting simple regression to training set
model=regressor.fit(x_train,y_train)
```

```python
# coefficient
model.coef_
```

```python
#intercept
model.intercept_
```

```python
#predicting the train results
y_pred=model.predict(x_train)
y_pred
```

```python
#predicting the test set results
y_pred=model.predict(x_test)
y_pred
```

```python
# Applying  logistic regression on same data
from sklearn.linear_model import LogisticRegression
```

```python
model2=LogisticRegression()
```

```python
# fitting the above data
model2.fit(x_train,y_train)
```

```python
model2.classes_
```

```python
#predicting the train data 
model2.predict(x_train)
```

```python
# predicting probability on train data
model2.predict_proba(x_train)
```

```python
#predicting on the test data
model2.predict(x_test)
```

```python
#predicting probability on test data
model2.predict_proba(x_test)
```
